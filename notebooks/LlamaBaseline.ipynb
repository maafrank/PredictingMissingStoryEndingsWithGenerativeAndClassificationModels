```python
'''from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load the Mistral model and tokenizer
model_name = "mistralai/Mistral-7B-Instruct-v0.3"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")'''

# Use a pipeline as a high-level helper
from transformers import pipeline
import torch

# Set up huggingface token
from google.colab import userdata
api_token = userdata.get('HF_TOKEN')

if api_token:
    from huggingface_hub import HfApi, HfFolder
    HfFolder.save_token(api_token)
    print("HuggingFace API token saved in userdata")
else:
    print("HuggingFace API token not found in userdata")


device = 0 if torch.cuda.is_available() else -1  # Set to 0 for GPU, -1 for CPU
pipe = pipeline("text-generation",
                model="meta-llama/Llama-3.1-8B-Instruct",
                device=device,
                torch_dtype=torch.float16,  # Use mixed precision
                )
```

```python
# Function to generate a fifth sentence
def generate_sentence(sentences):

    # Generate the fifth sentence
    generated_output = pipe(sentences,
                            max_length=5000,
                            num_return_sequences=1,
                            do_sample=True,
                            pad_token_id=2)

    # Extract the generated text from the model output
    generated_text = generated_output[0]["generated_text"][-1]["content"]
    #print(generated_text)

    # Split the output by newline to get individual fifth sentences
    fifth_sentences = [sentence.strip() for sentence in generated_text.split("\n") if sentence.strip()]
    return fifth_sentences

    return generated_output
```

```python
def get_message(examples_df, sentences_df):
    messages = [
        {"role": "user", "content": f"""I am going to give you a 4 sentence story and I need you to generate the 5th sentence. Please generate only 1 sentence."""},
        {"role": "assistant", "content": f"""Sure, I can help you generate the 5 sentence of your 4 sentence story. I will generate only 1 sentence."""},
      ]

    stories = ''
    for index, row in examples_df.iterrows():
      example_sentences = [row[f"sentence{i}"] for i in range(1, 5)]
      concatenated_sentences = " ".join(example_sentences)
      stories += concatenated_sentences + "\n"

    messages.append({"role": "user", "content": f"""{stories}"""})

    fifth_sentences = ''
    for index, row in examples_df.iterrows():
      example_sentences = [row[f"sentence{i}"] for i in range(5, 6)]
      concatenated_sentences = " ".join(example_sentences)
      fifth_sentences += concatenated_sentences + "\n"

    messages.append({"role": "assistant", "content": f"""{fifth_sentences}"""})

    stories = ''
    for index, row in sentences_df.iterrows():
      example_sentences = [row[f"sentence{i}"] for i in range(1, 5)]
      concatenated_sentences = " ".join(example_sentences)
      stories += concatenated_sentences + "\n"

    messages.append({"role": "user", "content": f"""{stories}"""})


    return messages
```

```python
batch_size = 5
examples_df = narrative_df.iloc[:batch_size].head(5)
data_df = narrative_df.iloc[batch_size:].head(40000)
```

```python
get_message(examples_df, data_df.head(batch_size))
```

```python
import pandas as pd
from tqdm import tqdm

# Prepare a list to store the results
results = {"sentence1": [],
           "sentence2": [],
           "sentence3": [],
           "sentence4": [],
           "sentence5": []}

# Process `data_df` in batches
#data_df = data_df.head(100)
for start_idx in tqdm(range(0, len(data_df), batch_size), desc="Generating Sentences in Batches"):
    # Select the current batch
    batch_df = data_df.iloc[start_idx:start_idx + batch_size]

    # Generate fifth sentences for the batch
    message = get_message(examples_df, batch_df)
    fifth_sentences = generate_sentence(message)

    # Append each row's data to the results dictionary
    for index, (j, row) in enumerate(batch_df.iterrows()):
        results["sentence1"].append(row["sentence1"])
        results["sentence2"].append(row["sentence2"])
        results["sentence3"].append(row["sentence3"])
        results["sentence4"].append(row["sentence4"])
        # Check if we have a generated sentence for this row
        if index < len(fifth_sentences):
            results["sentence5"].append(fifth_sentences[index])
        else:
            results["sentence5"].append("")

# Convert the results to a DataFrame
output_df = pd.DataFrame(results)

# Save the DataFrame to a CSV file with "mistral" in the filename
output_csv_path = os.path.join(data_dir, "generated_narratives_llama.csv")
output_df.to_csv(output_csv_path, index=False)

print(f"Results saved to {output_csv_path}")
```
