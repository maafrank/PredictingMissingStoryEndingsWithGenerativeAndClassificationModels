```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
pd.set_option('display.max_columns', None)
```

```python
# Define the directory containing the files
data_dir = '/content/drive/MyDrive/Colab Notebooks/Berkeley/266/FinalProject/Data/'

# Load each file into a dictionary of DataFrames
file_dict = {}
for filename in os.listdir(data_dir):
    if filename.endswith('.csv'):
        file_path = os.path.join(data_dir, filename)
        file_dict[filename] = pd.read_csv(file_path)
        print(f"{filename} loaded with shape: {file_dict[filename].shape}")
```

```python
for name, df in file_dict.items():
    print(f"\n{name}")
    display(df.info())
    display(df.head())
```

```python
# Combine Narrative Stories into one DataFrame
narrative_dfs = [df for name, df in file_dict.items() if 'ROCStories' in name]
combined_narrative_df = pd.concat(narrative_dfs, ignore_index=True)
print(f"Combined Narrative Stories DataFrame shape: {combined_narrative_df.shape}")

# Save combined narrative data
combined_narrative_path = os.path.join(data_dir, 'combined_narrative_stories.csv')
combined_narrative_df.to_csv(combined_narrative_path, index=False)
```

## Story Length Analysis

```python
# Calculate story length in words and add a new column
combined_narrative_df['story_length'] = combined_narrative_df.apply(
    lambda row: sum(len(str(row[f'sentence{i}']).split()) for i in range(1, 6)), axis=1
)

# Summary stats for story length
print("Story length statistics:")
print(combined_narrative_df['story_length'].describe())


# Histogram for story lengths
plt.figure(figsize=(10, 6))
plt.hist(combined_narrative_df['story_length'], bins=20, edgecolor='black')
plt.title('Distribution of Story Lengths')
plt.xlabel('Story Length (Number of Words)')
plt.ylabel('Frequency')
plt.show()
```

## Most Common Words or Phrases

```python
from collections import Counter
import itertools

# Combine all words in the story sentences into one list, ensuring each entry is a string
words = list(itertools.chain.from_iterable(
    str(sentence).split() for i in range(1, 6) for sentence in combined_narrative_df[f'sentence{i}']
))

# Get the most common words
word_counts = Counter(words)
print("Most common words:", word_counts.most_common(20))


# Get the 20 most common words
most_common_words = word_counts.most_common(20)
words, frequencies = zip(*most_common_words)  # Unzip into words and frequencies

# Plot a bar chart
plt.figure(figsize=(12, 6))
plt.bar(words, frequencies, color='skyblue')
plt.title('Top 20 Most Common Words in Stories')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

## Sentence Position Analysis

```python
for i in range(1, 6):
    sentence_col = f'sentence{i}'
    sentence_words = list(itertools.chain.from_iterable(
        combined_narrative_df[sentence_col].str.split().tolist()
    ))
    print(f"Most common words in {sentence_col}:", Counter(sentence_words).most_common(10))

# Set up the figure for multiple bar charts
fig, axes = plt.subplots(1, 5, figsize=(20, 6), sharey=True)
fig.suptitle("Most Common Words in Each Sentence Position")

# Loop through each sentence column, calculate the most common words, and plot them
for i in range(1, 6):
    sentence_col = f'sentence{i}'

    # Get a list of words in the current sentence column
    sentence_words = list(itertools.chain.from_iterable(
        combined_narrative_df[sentence_col].str.split().tolist()
    ))

    # Get the 10 most common words for the current sentence column
    common_words = Counter(sentence_words).most_common(10)
    words, frequencies = zip(*common_words)

    # Plot the bar chart for the current sentence position
    ax = axes[i-1]
    ax.bar(words, frequencies, color='skyblue')
    ax.set_title(f'{sentence_col.capitalize()}')
    ax.set_xticklabels(words, rotation=45)
    ax.set_xlabel("Words")
    if i == 1:
        ax.set_ylabel("Frequency")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## Sentence Length Analysis by Position

```python
import numpy as np

sentence_lengths = {
    f'sentence{i}': combined_narrative_df[f'sentence{i}'].apply(lambda x: len(str(x).split())).mean()
    for i in range(1, 6)
}
plt.figure(figsize=(8, 5))
plt.bar(sentence_lengths.keys(), sentence_lengths.values(), color='lightgreen')
plt.xlabel("Sentence Position")
plt.ylabel("Average Sentence Length (words)")
plt.title("Average Sentence Length by Position")
plt.show()
```

## Sentiment Analysis

```python
from textblob import TextBlob

# Calculate sentiment for each sentence position
for i in range(1, 6):
    sentence_col = f'sentence{i}'
    combined_narrative_df[f'sentiment_{sentence_col}'] = combined_narrative_df[sentence_col].apply(
        lambda x: TextBlob(str(x)).sentiment.polarity
    )

# Average sentiment by sentence position
avg_sentiments = [combined_narrative_df[f'sentiment_sentence{i}'].mean() for i in range(1, 6)]

# Plot sentiment by sentence position
plt.figure(figsize=(8, 5))
plt.plot(range(1, 6), avg_sentiments, marker='o', linestyle='-', color='coral')
plt.xlabel("Sentence Position")
plt.ylabel("Average Sentiment Polarity")
plt.title("Average Sentiment Polarity by Sentence Position")
plt.show()
```

## Bigram and Trigram Analysis

```python
from sklearn.feature_extraction.text import CountVectorizer

# Bigrams
vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words='english')
X = vectorizer.fit_transform([' '.join([str(row[f'sentence{i}']) for i in range(1, 6)]) for _, row in combined_narrative_df.iterrows()])
bigram_counts = Counter(dict(zip(vectorizer.get_feature_names_out(), X.sum(axis=0).A1)))
print("Most common bigrams:", bigram_counts.most_common(10))

# Trigrams
vectorizer = CountVectorizer(ngram_range=(3, 3), stop_words='english')
X = vectorizer.fit_transform([' '.join([str(row[f'sentence{i}']) for i in range(1, 6)]) for _, row in combined_narrative_df.iterrows()])
trigram_counts = Counter(dict(zip(vectorizer.get_feature_names_out(), X.sum(axis=0).A1)))
print("Most common trigrams:", trigram_counts.most_common(10))


# Extract most common bigrams and trigrams
top_bigrams = bigram_counts.most_common(10)
top_trigrams = trigram_counts.most_common(10)

# Separate the words and counts for plotting
bigram_words, bigram_frequencies = zip(*top_bigrams)
trigram_words, trigram_frequencies = zip(*top_trigrams)

# Plotting the bigrams
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)  # create a subplot for bigrams
plt.barh(bigram_words, bigram_frequencies, color='lightblue')
plt.gca().invert_yaxis()  # invert y-axis for better readability
plt.title('Top 10 Most Common Bigrams')
plt.xlabel('Frequency')
plt.ylabel('Bigrams')

# Plotting the trigrams
plt.subplot(1, 2, 2)  # create a subplot for trigrams
plt.barh(trigram_words, trigram_frequencies, color='lightgreen')
plt.gca().invert_yaxis()  # invert y-axis for better readability
plt.title('Top 10 Most Common Trigrams')
plt.xlabel('Frequency')
plt.ylabel('Trigrams')

plt.tight_layout()
plt.show()
```

## Word Cloud Visualization

```python
from wordcloud import WordCloud

for i in range(1, 6):
    sentence_col = f'sentence{i}'
    text = ' '.join(combined_narrative_df[sentence_col].dropna().astype(str).tolist())
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for {sentence_col.capitalize()}')
    plt.show()
```
